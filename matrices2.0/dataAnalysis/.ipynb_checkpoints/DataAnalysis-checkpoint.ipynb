```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = "../outputs/matrixAPP.out-size"
sizes = [50, 100, 200, 300, 400, 500, 600, 700, 800]
data = np.zeros([30, 9])
```

```python
for i, size in enumerate(sizes):
    f = open(path+str(size))
    for j, l in enumerate(f):
        data[j][i] = float(l);
```

```python
df = pd.DataFrame(data=data, columns=sizes)
df
```

```python
mean = df.mean()
mean
```

```python
mean.plot(xlabel="size (NxN)", ylabel="time", title="Medidas de rendimiento")
```

```python
mean.plot(kind="bar", xlabel="size (NxN)", ylabel="time", title="Medidas de rendimiento")
```

```python
df.to_csv(path+"results.csv", index=False)
```
