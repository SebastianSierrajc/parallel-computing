```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = "../outputs/"
file_name = "matrixAPP.out-size"
ext = ".txt"
sizes = [50, 100, 200, 300, 400, 500, 600, 700, 800, 1000, 1200, 2000, 4000]
data = np.zeros([30, len(sizes)])
```

```python
for i, size in enumerate(sizes):
    f = open(path+file_name+str(size)+ext)
    for j, l in enumerate(f):
        data[j][i] = float(l);
```

```python
df = pd.DataFrame(data=data, columns=sizes)
df
```

```python
mean = df.mean()
mean
```

```python
mean.plot(xlabel="size (NxN)", ylabel="time (us)", title="Medidas de rendimiento")
```

```python
mean.plot(kind="bar", xlabel="size (NxN)", ylabel="time (us)", title="Medidas de rendimiento")
```

```python
df.to_csv(path+"results.csv", index=False)
```
